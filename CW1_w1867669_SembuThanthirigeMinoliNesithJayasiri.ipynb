## Artificial Intelligence in Medicine


With the rapid advancement of Artificial Intelligence (AI) in the recent years many contributions have been made to the field of medicine. 
The use of image classification has proven to be of great use in the interpretation of medical images to help diagnose patient medical conditions.
AI has also contributed to the discovery of drugs, reducing the cost and time needed for physical experiments (Rajpurkar et al., 2022). Natural 
language processing is a branch of AI which involves analyzing human languages, studies have been carried out in this field to build chatbots that
are able to answer medical related questions and support patients. 

(Badža and Barjaktarović, 2020)carried out a study to classify brain tumors from MRI images. They introduced a Convolutional Neural Networks (CNN)
architecture to classify three types of tumors. The CNN consisted of 22 layers in total. Adam was the optimizer used with an initial learning rate 
of 0.0004 and a training batch size of 16. K-fold cross-validation had been used to test the model. The model achieved an accuracy of 96.5%. Once 
of the main highlights of this study is that the model does not need many resources for training and its execution is fast.

Another research was conducted by Ibrahim et al., 2021, for the classification of chest x-ray images to detect Covid-19 Pneumonia. They utilized the
pre-trained model AlexaNet for the classification. Publicly available datasets were used for this study. AlexaNet uses Relu as the activation function
in convolutional layers and Softmax activation function for classification. The input size of images AlexaNet was previously trained on is 227x227x3. 
Accuracy, specificity, and sensitivity were used to evaluate the model. 

(Dahiwade et al., 2019), proposed their research on predicting the disease according to the symptoms given as input using machine learning. The dataset 
was taken from UCI machine learning repository.  The dataset was trained on both KNN and CNN architecture. The CNN architecture gave an accuracy of 84.5% 
which proved to be higher than the accuracy given by the KNN algorithm. It also took less time to perform classification compared to KNN.

Fitriyani et al., 2019, explored the use of  an ensemble learning approach for the prediction of Diabetes and Hypertension. Four datasets were used for the
training process. Ensemble learning involves combining multiple models to form one single model to improve its performance. The stacking ensemble approach was 
used in this research. Multi-Layer Perceptron (MLP), Support Vector Machine (SVM) and Decision tree (DT) classifiers were used as base learners, and their output
was used as features for the Logistic Regression (LR) to form the final prediction. The model achieved an accuracy of 96.7%, 85.73%, 75.78% and 100% for the 
four datasets respectively. 

Divya, Priyasankari and Devi, n.d., proposed a medical chatbot built using Artificial Intelligence which can diagnose diseases based on the input provided by
the user. The chatbot consisted of three conversational phases which were, getting general information, the extraction of symptoms and then diagnosing potential
disease types. The string searching algorithm was used to extract symptoms then symptoms closest to the extracted strings are displayed to the user. 

Interesting discoveries have also been made in the field of drug discovery using AI. AI has helped predict the structure of proteins and also in understanding
drug activity. AI has been used in the Poly pharmacology industry in designing drug molecules. AI has also proved helpful in the prediction of yield and getting 
insights into reaction mechanisms in the chemical processes (Paul et al., 2020).

## Heart Disease Prediction 

### Ensemble Learning

(H. et al., 2016) proposed a technique to predict heart disease using Ensemble Learning. Dataset used was taken from the heart disease databases available in the UCI Machine Learning Repository. Four different databases were used for this research. The output data contained values from “0” to “4” where “0” represented the absence of heart disease and “1-4” represented different risk levels of heart disease. This was reclassified into classes of “1” and “0” indicating the absence and presence of heart disease respectively. 
The Adaptive Boosting Algorithm was used for the implementation of the model. It is an ensemble learning approach that combines outputs of other algorithms into a weighted sum to form one strong prediction model.
Several methods including, specificity, sensitivity, precision, probability of the misclassification error(PME), accuracy, f1 score and ROC Curve were used for the testing of the proposed model. An average accuracy of 85.2% was obtained for the four datasets.
The algorithm used in this research is sensitive to noise and outliers in data, therefore leads to overfitting. Research on methods to reduce the overfitting problem was kept as part of the future work in this research.


### Traditional Machine Learning 

(Shah et al., 2020) carried out research on techniques to predict the presence of heart disease using different tradition machine learning algorithms. Dataset used was obtained from the UCI Machine learning repository. The dataset contains 13 feature columns and 1 column for the classification result. 
The dataset was preprocessed and  trained using four different algorithms which include, Naïve Bayed Classifier, K-Nearest Neighbor(KNN), Decision tree and Random Forest Classifier. The models achieved an accuracy of 88.157%, 90.789%, 80.263% and 86.84% respectively. The KNN algorithm with k as 7, obtained the highest accuracy out of the four algorithms used.
The implementation of a more complex model with a combination of different models would be considered in future research on order to achieve a better prediction accuracy.  


### Neural Networks

(Nanekar, 2021)  conducted research on a technique to predict the presence of heart disease using a multilayer perceptron neural network. The Cleveland dataset was used for this research. The dataset contains 213 records with 13 attributes. The missing values in the dataset were filled with interpolation values. The dataset was split into three parts 70% for training, 15% for testing and 15% for validation. The output was designed to have two output types “1” for the presence of heart disease and “0” for the absence of heart disease. The Backpropagation Algorithm was used for the training of the dataset. The algorithm contains one input layer with 13 neurons, one hidden layer with 12 neurons and one output layer with 2 neurons. To further improve the performance of the system pruning was used. Accuracy, precision, and recall were used for the evaluation of the model. The proposed system achieved an accuracy of 95%.

```python
dataset = dataset.replace(to_replace='Presence',value=1)
```

```python
dataset = dataset.replace(to_replace='Absence',value=0)
```

```python
x = dataset.drop('Heart Disease', axis = 1)
y = dataset['Heart Disease']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

### Oversampling data due to class imbalance

```python
presence = sum(sum([y_train == 1]))
absence = sum(sum([y_train == 0]))

print("Presence:", presence)
print("Absence:", absence)
```

```python
from imblearn.over_sampling import RandomOverSampler

x_train, y_train = RandomOverSampler(sampling_strategy='minority').fit_resample(x_train, y_train)
```

```python
presence = sum(sum([y_train == 1]))
absence = sum(sum([y_train == 0]))

print("Presence:", presence)
print("Absence:", absence)
```

### Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Random Forest Classifier - Implementation and Evaluation

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf_train_acc = rf.score(x_train,y_train)

rf_pred = rf.predict(x_test)

confusion_matrix = metrics.confusion_matrix(y_test, rf_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

rf_accuracy = metrics.accuracy_score(y_test, rf_pred)
rf_precision = metrics.precision_score(y_test, rf_pred)
rf_recall = metrics.recall_score(y_test, rf_pred)
rf_f1_score = metrics.f1_score(y_test, rf_pred)
#metrics
print("The accuracy is :",round(rf_accuracy,2)*100,"%")
print("The precision is :",round(rf_precision,2)*100,"%")
print("The recall is :",round(rf_recall,2)*100,"%")
print("The f1-score is :",round(rf_f1_score,2)*100,"%")
```

# Logistic Regression Classifier - Implementation and Evaluation

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

LR = LogisticRegression()
LR.fit(x_train, y_train)
lr_train_acc = LR.score(x_train,y_train)

LR_pred = LR.predict(x_test)

confusion_matrix = metrics.confusion_matrix(y_test, LR_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

lr_accuracy = metrics.accuracy_score(y_test, LR_pred)
lr_precision = metrics.precision_score(y_test, LR_pred)
lr_recall = metrics.recall_score(y_test, LR_pred)
lr_f1_score = metrics.f1_score(y_test, LR_pred)
#metrics
print("The accuracy is :",round(lr_accuracy,2)*100,"%")
print("The precision is :",round(lr_precision,2)*100,"%")
print("The recall is :",round(lr_recall,2)*100,"%")
print("The f1-score is :",round(lr_f1_score,2)*100,"%")
```

# K-Nearest Neighbour Classifier - Implementation and Evaluation

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

accuracy_scores = []
for n in range(1,21):
    KNN = KNeighborsClassifier(n_neighbors = n)
    KNN.fit(x_train, y_train)
    KNN_predict = KNN.predict(x_test)
    accuracy_scores.append(metrics.accuracy_score(y_test, KNN_predict))

max_score = accuracy_scores[0]
n = 1
k = 0
for score in accuracy_scores:
    if score > max_score:
        max_score = score
        k = n
    n = n+1 

KNN = KNeighborsClassifier(n_neighbors = k)
KNN.fit(x_train, y_train)
KNN_pred = KNN.predict(x_test)
knn_train_acc = KNN.score(x_train,y_train)

confusion_matrix = metrics.confusion_matrix(y_test, KNN_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

knn_accuracy = metrics.accuracy_score(y_test, KNN_pred)
knn_precision = metrics.precision_score(y_test, KNN_pred)
knn_recall = metrics.recall_score(y_test, KNN_pred)
knn_f1_score = metrics.f1_score(y_test, KNN_pred)
#metrics
print("The accuracy is :",round(knn_accuracy,2)*100,"%")
print("The precision is :",round(knn_precision,2)*100,"%")
print("The recall is :",round(knn_recall,2)*100,"%")
print("The f1-score is :",round(knn_f1_score,2)*100,"%")
```

# Naive Bayes - Implementation and Evaluation

```python
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

nb = GaussianNB()
nb.fit(x_train, y_train)
nb_pred = nb.predict(x_test)
nb_train_acc = nb.score(x_train,y_train)

confusion_matrix = metrics.confusion_matrix(y_test, nb_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

nb_accuracy = metrics.accuracy_score(y_test, nb_pred)
nb_precision = metrics.precision_score(y_test, nb_pred)
nb_recall = metrics.recall_score(y_test, nb_pred)
nb_f1_score = metrics.f1_score(y_test, nb_pred)
#metrics
print("The accuracy is :",round(nb_accuracy,2)*100,"%")
print("The precision is :",round(nb_precision,2)*100,"%")
print("The recall is :",round(nb_recall,2)*100,"%")
print("The f1-score is :",round(nb_f1_score,2)*100,"%")
```

# Decision Tree - Implementation and Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_test)
dt_train_acc = dt.score(x_train,y_train)

confusion_matrix = metrics.confusion_matrix(y_test, dt_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

dt_accuracy = metrics.accuracy_score(y_test, dt_pred)
dt_precision = metrics.precision_score(y_test, dt_pred)
dt_recall = metrics.recall_score(y_test, dt_pred)
dt_f1_score = metrics.f1_score(y_test, dt_pred)
#metrics
print("The accuracy is :",round(dt_accuracy,2)*100,"%")
print("The precision is :",round(dt_precision,2)*100,"%")
print("The recall is :",round(dt_recall,2)*100,"%")
print("The f1-score is :",round(dt_f1_score,2)*100,"%")
```

# Ensemble Learning - Stacking - Implementation and Evaluation

```python
from sklearn.ensemble import StackingClassifier
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

estimators = [
    ('KNN',KNN),
    ('DT',dt),
    ('lr',LR),
    ('NB',nb),
    ('RF',rf)]

stack_model = StackingClassifier(estimators,final_estimator=LogisticRegression())

stack_model.fit(x_train,y_train)

stack_pred = stack_model.predict(x_test)
stack_train_acc = stack_model.score(x_train,y_train)

confusion_matrix = metrics.confusion_matrix(y_test, stack_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

stack_accuracy = metrics.accuracy_score(y_test, stack_pred)
stack_precision = metrics.precision_score(y_test, stack_pred)
stack_recall = metrics.recall_score(y_test, stack_pred)
stack_f1_score = metrics.f1_score(y_test, stack_pred)
#metrics
print("The accuracy is :",round(stack_accuracy,2)*100,"%")
print("The precision is :",round(stack_precision,2)*100,"%")
print("The recall is :",round(stack_recall,2)*100,"%")
print("The f1-score is :",round(stack_f1_score,2)*100,"%")
```

# Ensemble Learing - Voting - Implementation and Evaluation

```python
from sklearn.ensemble import VotingClassifier
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

vot_model = VotingClassifier([
        ('KNN', KNN),
        ('DT',dt),
        ('lr',LR),
        ('GB', nb),
         ('RF', rf)
],
 voting = 'soft'
                                 )

vot_model.fit(x_train, y_train)

vot_train_acc = vot_model.score(x_train,y_train)
vot_pred = vot_model.predict(x_test)

confusion_matrix = metrics.confusion_matrix(y_test, vot_pred)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["No Heart Disease", "Heart Disease"])

cm_display.plot()
plt.show()

vot_accuracy = metrics.accuracy_score(y_test, vot_pred)
vot_precision = metrics.precision_score(y_test, vot_pred)
vot_recall = metrics.recall_score(y_test, vot_pred)
vot_f1_score = metrics.f1_score(y_test, vot_pred)
#metrics
print("The accuracy is :",round(vot_accuracy,2)*100,"%")
print("The precision is :",round(vot_precision,2)*100,"%")
print("The recall is :",round(vot_recall,2)*100,"%")
print("The f1-score is :",round(vot_f1_score,2)*100,"%")
```

<span style="font-size: 16px;">1.TP (True Positive): TP represents the number of patients who have been properly classified to have heart disease.<br>
2.TN (True Negative): TN represents the number of correctly classified patients who do not have heart disease.<br>

4.FN (False Negative): FN represents the number of patients misclassified as healthy but actually they have heart disease.<br></span>

<span style="font-size: 16px;">The performance metrics accuracy, precision,recall and f1-score are calculated based on TP, TN, FP and FN values.</span>

<span style="font-size: 16px;">1. Accuracy - It is the ratio of correctly classified instances to the total number of instances.<br>
2. Precision - It is the ratio of correctly classified patients with heart disease to the total number of patients with heart disease.<br>
3. Recall - It is the ratio of correctly classified patients without heart disease to the total number of patients without heart disease.<br>
4. F1 Score - It gives the equilibrium between precision and recall.<br></span>

## Time Complexities of training

#### Random Forest - O(n* m * log(m))
n - number of trees in the forest

m - number of features in the dataset

#### Logistic Regression - O(n * d )
n - number of training samples

d - number of features

#### K-Nearest Neighbour - O(n * d)
n - number of training samples

d - number of features

#### Naive Bayes - O(n* d *c)
n - number of training samples

d - number of features

c - number of classes

#### Decision Tree - O(n * d * log(n))
n - number of training samples

d - number of features

log(n) - depth of the tree

#### In Stacking Ensemble Learning method the time complexity is the sum of all time complexities of base models + the time complexity of the meta-model, which is the model that combines the predictions of all models.

#### In Voting Ensemble Learning method the time complexity is the sum of the time complexities of the individual models + the time complexity of the aggregration on the model preditions (O(n)).
